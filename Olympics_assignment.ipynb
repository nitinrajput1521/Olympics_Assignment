```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer_olympics.csv")
```

```python
df.head()
```

```python
len(df["City"].unique())  # finding number of cities
```

```python
data = []
for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Gold']).sort_values(by='Gold', ascending=False).head().plot(x = 'Sport', y = 'Gold', color='blue', 
                                                                                                  kind = 'bar', figsize = (10,5))
```

```python
data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar',color="green" ,figsize = (10,5))
```

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete , len(df[df['Athlete']  == Athlete])])

pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (10,5))
```

```python
plt.figure(figsize=(15, 10))
tops=df[df['Medal']=='Gold'].Athlete.value_counts().head()
tops.plot(kind='bar',color="Pink")
plt.title('player has won most number Gold Medals of medals')
plt.show()
```

```python
data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])

pd.DataFrame(data,columns = ['Event','Athlete']).sort_values(by='Athlete',ascending=False).head().plot(x = 'Event', y = 'Athlete', kind = 'bar',color="Orange", figsize = (10,5))
```

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete , len (df[(df.Gender == "Women") & (df.Medal == "Gold")])])

pd.DataFrame(data,columns = ['Athlete','freq']).sort_values(by='freq',ascending=False).head().plot(x = 'Athlete', y = 'freq', kind = 'bar',color="cyan", figsize = (10,5))
```
